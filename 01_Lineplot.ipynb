Line Plot

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
p=sns.load_dataset("iris")
p

#draw line plot(line plot is always between two numeric variables)
sns.lineplot(x='sepal_length', y='sepal_width', data=p)
plt.show()
```

Adding title

```python

#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
p=sns.load_dataset("iris")
p

#draw line plot(line plot is always between two numeric variables)
sns.lineplot(x='sepal_length', y='sepal_width', data=p)
plt.title('Flower Plot')
plt.show()
```

Adding Limits

```python
import seaborn as sns
import matplotlib.pyplot as plt

p=sns.load_dataset("iris")

sns.lineplot(x='sepal_length',y='sepal_width',data=p)
plt.title('Flower Plot')
plt.xlim(2)
plt.ylim(1)
plt.show()
```

Set Styles

```python
import seaborn as sns
import matplotlib.pyplot as plt

p=sns.load_dataset("iris")

sns.set_style("darkgrid")
#sns.set_style(style=None,rc=None)
sns.lineplot(x='sepal_length',y='sepal_width',data=p)
plt.title('Flower Plot')

plt.show()
```

Size of Figure

```python
import seaborn as sns
import matplotlib.pyplot as plt

p=sns.load_dataset("iris")

sns.set_style("darkgrid")
#sns.set_style(style=None,rc=None)
#change fig. size
plt.figure(figsize=(4,2))
sns.lineplot(x='sepal_length',y='sepal_width',data=p)

plt.title('Flower Plot')




plt.show()
```
