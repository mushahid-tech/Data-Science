How to manage individual color for hue variables

```python
import seaborn as sns
import matplotlib.pyplot as plt

p=sns.load_dataset('tips')
sns.set(style="darkgrid")

#boxplot of one variable
sns.boxplot(x='tip',y='day',hue='smoker',data=p,palette={'Yes':'red','No':'green'})

plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np

p=sns.load_dataset('titanic')
p.head()
```

```python
sns.boxplot(x='survived', y='age',data=p)
```

Interquartile range-From lower portion of the box to the upper portion of the box

Q1

Q2:Center= Median

Q3

Bar at the top: Maximum non outlier data

Bar at the bottom: Minimum non outlier data

Dots Above the bar at the top: Outliers(Missing data) 

```python
#showmeans
sns.boxplot(x='survived', y='age',showmeans=True,data=p)
```

```python
import matplotlib.pyplot as plt
sns.boxplot(x='survived', y='age',showmeans=True,meanprops={'marker':'+','markersize':'10','markeredgecolor':'red'},data=p)

#show labels
plt.xlabel('How many survived',size=14,weight='bold'),
plt.ylabel('Age(years',size=14,weight='bold'),
plt.title('Boxplot',size=14,weight='bold')
```

```python
sns.FacetGrid(p)
```
