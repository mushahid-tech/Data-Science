```python
import seaborn as sns
import matplotlib.pyplot as plt

p=sns.load_dataset("iris")

sns.set_style("darkgrid")
#sns.set_style(style=None,rc=None)
#change fig. size

sns.barplot(x='species',y='sepal_width',data=p)

plt.title('Flower Plot')




plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

p=sns.load_dataset("iris")

sns.set_style("darkgrid")
#sns.set_style(style=None,rc=None)
#change fig. size
plt.figure(figsize=(4,2))
sns.lineplot(x='sepal_length',y='petal_width',data=p)

plt.title('Flower Plot')




plt.show()
```
